```python
# 1 ile 100 arasındaki sayılardan 7'ye bölünenleri bulan Python kodu
sayi_dizisi=[]
sayi_dizisi2=[]
sayi=1
while sayi<=100:
  if sayi%7==0:
    sayi_dizisi.append(sayi)#append ekleme fonksiyonu
  elif sayi%7!=0:
    sayi_dizisi2.append(sayi)
  sayi+=1
print("1 ile 100 arasındaki 7'ye tam bölünen sayılar: ",sayi_dizisi)
print("Toplam sayı addedi: ",len(sayi_dizisi))
print("1 ile 100 arasındaki 7'ye tam bölünmeyen sayılar: ",sayi_dizisi2)
print("Toplam bölünmeyen sayı addedi: ",len(sayi_dizisi2))
```

```python
#kullanıcıdan başlangıç değeri alma
while True:
  try:
    baslangıc_degeri=int((input("Başlangıç değeri giriniz: ")))
    if baslangıc_degeri<0:
      baslangıc_degeri=abs(baslangıc_degeri)
      print("Negatif sayı girdiniz.Pozitife dönüştürdük. Başlangıç değeri: ",baslangıc_degeri)
    break
  except ValueError:
    print("Lütfen geçerli bir sayı giriniz...")
#kullanıcıdan son değer alma
while True:
  try:
    bitis_degeri=int((input("Bitiş değeri giriniz: ")))
    if bitis_degeri<0:
      bitis_degeri=abs(bitis_degeri)
      print("Negatif sayı girdiniz.Pozitife dönüştürdük. Bitiş değeri: ",bitis_degeri)
    break
  except ValueError:
    print("Lütfen geçerli bir sayı giriniz...")
#Başlangıç ve bitiş arasındaki çift sayıları bulma ve toplama
cift_sayilar=[]
cift_sayilarin_toplamı=0
while baslangıc_degeri<=bitis_degeri:
  if baslangıc_degeri%2==0:
   cift_sayilar.append(baslangıc_degeri)
   cift_sayilarin_toplamı +=baslangıc_degeri
  baslangıc_degeri+=1
#çift sayıları ve toplammını yazdırma
print(f"{baslangıc_degeri} ile {bitis_degeri} arasındaki sayılar: ")
for num in cift_sayilar:
  print(num,end=" ")
print(f"\nÇift sayıların toplamı: {cift_sayilarin_toplamı}")
```

```python
# Önceden tanımlalan sesli harf listesinde girilen metnin karşılaştırılması yapılarak sesli ve sessiz harf kontrolünün yapımı
sesli_harfler = 'aeıioöuüAEIİOÖUÜ'
kelime=input("Bir kelime giriniz: ")
#sesli ve sessiz harf sayısı
sesli_harf=0
sessiz_harf=0

#isalpha() ile bir stringde yalnızca harf olup olmadığını kontrol etmek için kullanılır.

#Aşağıdaki kod blogunda geçen "isalpha" harf dışı bir karakter ile karşılaşınca "false" döndürmektedir.
for harf in kelime:
  if harf.isalpha():
    if harf in sesli_harfler:
     sesli_harf += 1
    else:
     sessiz_harf += 1
print(kelime,"kelimesinde",sesli_harf,"adet sesli harf ve",sessiz_harf,"sessiz harf vardır.")
```
