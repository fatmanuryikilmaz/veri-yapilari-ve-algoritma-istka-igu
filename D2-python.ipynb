```python
# Dictionary - Sözlük Örneği -> Öğrencilerin notlarını tutan bir sözlük
notlar = {
    "Ali": 85,
    "Veli": 90,
    "Ayşe": 78
}

notlar["Fatma"] = 92 #öğrenci ekleme

notlar["Ali"] = 88 #öğrencinin notunu güncelleme

print(notlar)
```

```python
#set-küme örneği=>yazılım etkinliğine katılanların oluşturduğu kümeler
kodyazanlar= {"Ahmet","Mehmet","Ayşe"}
kodyazmayanlar= {"Ayşe","Fatma","Veli"}
#kafasına göre takılanlar
ortak=kodyazanlar.intersection(kodyazmayanlar)
print("kafasına göre takılan:",ortak)

#kod çalışması yapmayanlar
kod_yazmayanlar=kodyazmayanlar.difference(kodyazanlar)
print("kod çalışması yapmayanlar:",kod_yazmayanlar)

#yazılım eğitimi öğrenci imza listesi
imza_listesi=kodyazanlar.union(kodyazmayanlar)
print("katılımcılar:",imza_listesi)

```
